```python
def combine_features(row):
    str_user_id = str(row['user_id'])
    str_post_id = str(row['post_id'])
    str_title = str(row['title'])
    str_content = str(row['content'])
    
    return str_user_id +' '+ str_title +' '+str_content+' '+str_post_id
```

```python
for feature in features:
    df[feature] = df[feature].fillna('') #filling all NaNs with blank string
    df['combined_features'] = df.apply(combine_features,axis=1) #applying combined_features() method over each rows of dataframe and storing the combined string in “combined_features” column
```

```python
cv = CountVectorizer()
```

```python
count_matrix = cv.fit_transform(df["combined_features"])
```

```python
cosine_sim = cosine_similarity(count_matrix)
```

```python
print(cosine_sim)
```

```python
def get_title_from_index(id):
    for i in range(len(df.post_id)):
        if id != df.post_id[i]:
            return id, 'not in list'
        else:
#             return id, 'in llist'
            return df[df.post_id == id]['title'].values[0]
def get_index_from_title(title):
    return df[df.title == title]['post_id'].values[0]
```

```python
user_likes = 'This is lucid'
post_index = get_index_from_title(user_likes)
# print (post_index)
post_index = int(post_index)
similar_posts = list(enumerate(cosine_sim[post_index])) #accessing the row corresponding to given movie to find all the similarity scores for that movie and then enumerating over it
sorted_similar_posts = sorted(similar_posts,key=lambda x:x[1],reverse=True)[1:]
```
