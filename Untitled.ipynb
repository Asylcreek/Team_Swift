```python
# Read CSV files
post = pd.read_csv('posts.csv')
following = pd.read_csv('following.csv')
notification = pd.read_csv('notifications.csv')
user = pd.read_csv('users.csv')
```

```python
# Drop unnecessary fields from the table
post = post.drop(columns = ["action","tags","slug","created_at","updated_at","image","status_id","post_id"])
post.head()
```

```python
# Drop unnecessary fields from the table
notification = notification.drop(columns = ['parent_comment_id','comment','sender_id','user_id','status','created_at','updated_at'])
```

```python
#Merge table post and notification together
df = pd.merge(post, notification, on='id')
```

```python
df.drop(columns = ['id'])
```

```python
# Select importtant features
features = ['user_id','title','content','post_id','action','type']
```

```python
df.head(50)
```

```python
def combine_features(row):
    str_user_id = str(row['user_id'])
    str_post_id = str(row['post_id'])
    return str_user_id +' '+ row['title']+' '+row['content']+' '+str_post_id+' '+row['action']+' '+row['type']
```

```python
for feature in features:
    df[feature] = df[feature].fillna('') #filling all NaNs with blank string
    df['combined_features'] = df.apply(combine_features,axis=1) #applying combined_features() method over each rows of dataframe and storing the combined string in “combined_features” column
```

```python
cv = CountVectorizer()
```

```python
count_matrix = cv.fit_transform(df["combined_features"])
```

```python
cosine_sim = cosine_similarity(count_matrix)
```

```python
print(cosine_sim)
```

```python
def get_title_from_index(id):
    return df[df.id == id]['title'].values[0]
def get_index_from_title(title):
    return df[df.title == title]['id'].values[0]
```

```python
user_likes = 'HTML BEGINS HERE'
post_index = get_index_from_title(user_likes)
similar_posts = list(enumerate(cosine_sim[post_index])) #accessing the row corresponding to given movie to find all the similarity scores for that movie and then enumerating over it
sorted_similar_posts = sorted(similar_posts,key=lambda x:x[1],reverse=True)[1:]
```
